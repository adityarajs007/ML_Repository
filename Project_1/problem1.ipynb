```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#read csv
df = pd.read_csv('heart1.csv')
df
```

```python
#check for null values
def has_null_values(df):
    return df.isnull().values.sum() == 0

if has_null_values(df):
    print('no missing nor null values in data\n')
else:
    print('data has missing or nul values\n')
```

```python
#glance the data
df.describe()
```

```python
#covariance matrix
covariance_mat = df.cov()
covariance_mat
```

```python
#correlation matrix
correlation_mat = df.corr()
correlation_mat
```

```python
#to visualize histogram representation of data
df.hist(bins=30, figsize=(15,10))
plt.show()
```

```python
#heat map
plt.figure(figsize=(10,8))
sns.heatmap(correlation_mat, annot=True, cmap='coolwarm', fmt='.2f')
plt.show()
```

```python
#detect outliers
df.plot(kind='box', subplots=True, layout=(4, 4), figsize=(15,10))
```

```python
#Scatter plot
df.plot.scatter(x='age', y='rbp')
```

```python
#Baar plots
df['sex'].value_counts().plot(kind='bar')
```

```python
# VIF Variance Inflation Factor (VIF)
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = df.columns
vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
vif_data.sort_values(by='VIF', ascending=False)
```

```python
from scipy import stats
import numpy as np

# Calculate z-scores
z_scores = np.abs(stats.zscore(df.select_dtypes(include=[np.number])))
outliers = (z_scores > 3).sum(axis=0)
print("Outliers per column:", outliers)
```
